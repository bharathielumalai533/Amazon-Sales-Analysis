```python
plt.figure(figsize=(10,5))

plt.hist(df["actual_price"], bins=30)

plt.title("Distribution of Actual Product Prices")
plt.xlabel("Actual Price (₹)")
plt.ylabel("Number of Products")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(df["discount_percentage"], bins=20)

plt.title("Discount Percentage Distribution")
plt.xlabel("Discount (%)")
plt.ylabel("Products")

plt.show()
```

```python
top10 = (
    df[["product_name", "actual_price"]]
    .sort_values(by="actual_price", ascending=False)
    .head(10)
)
```

```python
plt.figure(figsize=(12,6))

sns.barplot(
    data=top10,
    x="actual_price",
    y="product_name"
)

plt.title("Top 10 Most Expensive Products")

plt.show()
```

```python
top_rating = (
    df[["product_name", "rating"]]
    .sort_values(by="rating", ascending=False)
    .head(10)
)
```

```python
plt.figure(figsize=(12,6))

sns.barplot(
    data=top_rating,
    x="rating",
    y="product_name"
)

plt.title("Top Rated Products")

plt.show()
```

```python
category = (
    df["category"]
    .value_counts()
    .head(10)
)
```

```python
plt.figure(figsize=(12,6))

sns.barplot(
    x=category.values,
    y=category.index
)

plt.title("Top 10 Categories")

plt.show()
```

```python
plt.figure(figsize=(8,6))

sns.heatmap(
    df[
        [
            "actual_price",
            "discounted_price",
            "discount_percentage",
            "rating",
            "rating_count"
        ]
    ].corr(),
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Heatmap")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(df["rating"], bins=10)

plt.title("Distribution of Product Ratings")
plt.xlabel("Rating")
plt.ylabel("Count")

plt.show()
```
